## Exploring the dataset of metadata

```python
connPh = sqlite3.connect("./NCEDC_PHASE/ncedc_sql")
tbl = pd.read_sql_query("SELECT * FROM ncedc_sql", connPh);
```

```python
# checking the table
tbl.head()
```

```python
# checking the overal statistics of data
tbl.describe()
```

```python
# magnitude distribution of data
import matplotlib.pyplot as plt
tbl['magnitude'].plot(kind='hist', bins= 50, normed= True)
plt.title('Hist')
plt.show()

# its culumative version
tbl.plot(y='magnitude', kind='hist', bins= 50, cumulative=True, normed= True)
plt.title('Cumulative dist')
plt.xlabel("Magnitude")
plt.show()
```

```python
# distance/magnitude distribution
tbl.plot(kind='scatter', x='distance', y='magnitude')
plt.show()
```

```python
# the largest event in the dataset
print(tbl['magnitude'].max())
```

```python
# just a rough look at the spacial distribution of data
plt.scatter(x= tbl['event_lon'], y =tbl['event_lat'], s = tbl['magnitude']*10, alpha = 0.8)
plt.style.use('ggplot')
plt.show()
```
